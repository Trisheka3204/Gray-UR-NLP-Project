```python
%%R 


i11_clusters %>% 
  add_count(i11_topic, name = "i11_topic_n") %>% 
  select(I11_cluster, i11_topic, i11_topic_n) %>% 
  arrange(desc(i11_topic_n))
```

```python
%%R 


i11_clusters %>% 
  group_by(i11_topic) %>%
  summarize(i11_topic_n = n()) %>% 
  ggplot(aes(x = fct_reorder(i11_topic, i11_topic_n), y = i11_topic_n)) +
  geom_col() +
  coord_flip() +
  theme_light() +
  labs(x = "Topic",
       y = "Number of comments",
       title = "Responses to I11 - What did  1216 instructor do during the transition to online learning that helped you to stay engaged in the course?") +
  theme(plot.title = element_text(hjust = 0.5, size = 12))
```
