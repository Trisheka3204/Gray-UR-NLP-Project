```python
import pandas as pd
import re
import spacy
import neuralcoref
import en_core_web_lg

nlp = en_core_web_lg.load()
neuralcoref.add_to_pipe(nlp)
```

```python
# this cell keeps creating a runtime error
doc1 = nlp("The dog is Fred's. It is loved by him. He walks it every day")
print(doc1._.coref_clusters)
print(doc1._.coref_resolved)

```

```python
doc2 = nlp('Angela lives in Boston. That city is her favorite.')
for ent in doc2.ents:
    print(ent._.coref_cluster)
```

```python
doc3 = nlp("The students used their scholarships to pay for college tuition it cost a lot to them they spent four years at the school.")
#doc3 = nlp("I wanted to help engineer things. They would help a lot.")
print(doc3._.coref_clusters)
print(doc3._.coref_resolved)
```

```python
doc4 = nlp("The students read the book in class. They enjoyed it.")
```

```python
resolved_str = doc4._.coref_resolved
res_doc = nlp(resolved_str)
for sent in res_doc.sents:
  print(sent)
```

```python
for sentence in res_doc.sents:
  for token in sentence:
    print(token.text, token.pos_, token.dep_, token.head)
```

```python
for i, sentence in enumerate(res_doc.sents):
  nsubj = None
  root = None
  dobj = None
  for token in sentence:
    print(i, token)
    if token.dep_ == 'nsubj':
      nsubj = token
      print(f"The nsubj is: {token}")
    if token.dep_ == "ROOT":
      root = token
      print(f"The ROOT is: {token}")
    if token.dep_ == "dobj":
      dobj = token
      print(f"The dobj is: {token}")
  
  if nsubj and root and dobj:
    print(nsubj, root, dobj)
```
